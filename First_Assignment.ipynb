1.What is Python and Why is it So Popular?
 - Python is a high-level programming language widely used in Data Science for analyzing data, building models, and creating intelligent applications.
It is popular because it has powerful libraries like NumPy, Pandas, Matplotlib, Scikit-learn, and TensorFlow, which make data cleaning, visualization, and machine learning much easier. 2.What is Interpreter in python?
 - An interpreter in Python is a program that reads and executes Python code line by line.
In Data Science, this is helpful because we can test small parts of code quickly - such as data cleaning steps, visualizations or model training - without compiling the whole program. This makes experiments faster and easier while working with datasets.

3.What are Predefined Keywords in Python?
 - Predefined keywords in Python are special words that have fixed meanings and are reserved for the Python language.
We cannot use them as variable names because they control the structure of a program - such as loops, conditions, and functions.
Common examples include: if, else, for, while, def, return, import, True, False ect.

4. Can keywords be used as variable names?
 - No, keywords cannot be used as variable names in Python.
They are reserved words with special meanings, such as if, for, while, import, True, False, and Python uses them to control the program's structure.


5.What is mutability in Python?
 - Mutability in Python means the ability of an object to change its value after it is created.Mutable objects can be modified, while immutable objects cannot be changed once created.
- For example, lists and dictionaries are mutable, but tuples and strings are immutable.

6. Why are lists mutable, but tuples are immutable?
- Lists are mutable because Python allows us to change, add, or remove their elements after creation.
- Tuples are immutable because Python locks their values once they are created.
This makes tuples safer and faster for storing fixed data — such as coordinates, or constant information that should not change.


7. What is the difference between “ == ” and “ is ” operator in Python?
- == Operator checks if two values are equal, even if they are stored in different memory locations.
- is--Operator checks if two variables refer to the same object in memory.
It is about identity, not value.
- This is useful when we want to confirm if two variables point to the exact same data object.


8. What are logical operators in Python?
- Logical operators in Python are used to combine multiple conditions and return True or False.
There are three types.
- and - returns True only if both conditions are true
- or - returns True if at least one condition is true
- not - reverses the result of a condition


9. What is type casting in Python?
- Type casting in Python means converting one data type into another, such as converting a string to an integer or a float to an integer. We convert them into proper types to perform calculations, analysis, and visualizations.


 10.What is the difference between implicit and explicit type casting?
 <br>Implicit Type Casting
- Python automatically converts one data type to another.
- It happens without the programmer writing anything.
- Example: converting an integer to a float during calculations.
<br> Explicit Type Casting
- The programmer manually converts one data type to another using functions like int(), float(), str().
- Used when we need full control over data formats.




11. What is the purpose of conditional statements in Python?
- The purpose of conditional statements in Python is to make decisions in a program based on certain conditions.
They allow the code to execute different actions depending on whether a condition is True or False.


12. How does the elif statement work?
<br>The elif statement is used when you want to check multiple conditions, one after another.
<br> If the first if condition is False, Python checks the elif condition.
<br>If that is also False, it moves to the next elif or finally to else.


13. What is the difference between for and while loops?
<br> For loop
- Used when you know the exact number of repetitions.
- In Data Science, it is commonly used to loop through rows, columns, or arrays.
<br>While loop
- Used when you don’t know how many times the loop will run.
- It continues until a specific condition becomes False.


14. Describe a scenario where a while loop is more suitable than a for loop.
- A while loop is more suitable when you don't know how many times the loop needs to run, and the repetition depends on a condition instead of a fixed range.

```python
# 14. Write a program to reverse a string using a while loop.

#input
string = input("Enter a string: ")

reverse_str = ""
i = len(string) - 1

while i >= 0:
    reverse_str += string[i]
    i -= 1

print("Reversed string:", reverse_str)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

fact=1
f=1
num=int(input("Enter the number of factorial: "))
while f <= num:
  fact= fact * f
  f += 1
print(fact)
```
